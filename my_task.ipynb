```python
import pandas as pd

# Read the input CSV file into a DataFrame
df = pd.read_csv('input.csv')

# Initialize an empty list to store transformed data
transformed_data = []

# Iterate through each row of the DataFrame
for index, row in df.iterrows():
    employee_code = row['Employee Code']
    manager_employee_code = row['Manager Employee Code']

    # Extract compensation data
    last_compensation = None
    compensation = None
    last_pay_raise_date = None
    variable_pay = None

    if pd.notna(row['Compensation']):
        last_compensation = row['Compensation']
    if pd.notna(row['Compensation 1']):
        compensation = row['Compensation 1']
        last_pay_raise_date = row['Compensation 1 date']
    if pd.notna(row['Compensation 2']):
        compensation = row['Compensation 2']
        last_pay_raise_date = row['Compensation 2 date']

    # Extract engagement data
    performance_rating = None
    engagement_score = None

    if pd.notna(row['Review 1']):
        performance_rating = row['Review 1']
    if pd.notna(row['Review 2']):
        performance_rating = row['Review 2']

    if pd.notna(row['Engagement 1']):
        engagement_score = row['Engagement 1']
    if pd.notna(row['Engagement 2']):
        engagement_score = row['Engagement 2']

    effective_date = row['Date of Joining']
    end_date = row['Date of Exit']

    # Create separate rows for each historical record
    transformed_data.append([employee_code, manager_employee_code, last_compensation, compensation,
                             last_pay_raise_date, variable_pay, effective_date, end_date,
                             performance_rating, engagement_score])

# Convert the list of lists to a DataFrame
transformed_df = pd.DataFrame(transformed_data, columns=['Employee Code', 'Manager Employee Code',
                                                         'Last Compensation', 'Compensation',
                                                         'Last Pay Raise Date', 'Variable Pay',
                                                         'Effective Date', 'End Date',
                                                         'Performance Rating', 'Engagement Score'])

# Write the transformed data to a new CSV file
transformed_df.to_csv('transformed_employee_data.csv', index=False)
```
